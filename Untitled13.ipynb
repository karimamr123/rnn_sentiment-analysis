```python
# ----------------------------
# Enhanced Schema Analysis
# ----------------------------
def analyze_schema(schema_chunks):
    """Extract tables, columns, relationships and keywords from schema dynamically"""
    schema_info = {
        'tables': {},
        'relationships': [],
        'keywords': set(),
        'data_types': set(),
        'common_columns': set(),
        'primary_keys': {},
        'foreign_keys': {}
    }

    # Common SQL data types to recognize
    data_types = [
        'INT', 'VARCHAR', 'TEXT', 'DATE', 'DATETIME', 'TIMESTAMP',
        'DECIMAL', 'ENUM', 'BOOLEAN', 'FLOAT', 'DOUBLE'
    ]

    # List of common SQL keywords to filter out
    sql_stopwords = [
        'table', 'column', 'columns', 'primary', 'key', 'foreign', 'references',
        'auto_increment', 'default', 'unique', 'index', 'constraint'
    ]

    # Add NLTK stopwords
    english_stopwords = set(stopwords.words('english'))

    for chunk in schema_chunks:
        # Extract table name
        table_match = re.search(r"Table: (\w+)", chunk)
        if table_match:
            table_name = table_match.group(1)
            schema_info['tables'][table_name] = {'columns': []}
            schema_info['keywords'].add(table_name.lower())

            # Extract singular form of table name as keyword too
            if table_name.endswith('s'):
                schema_info['keywords'].add(table_name[:-1].lower())

            # Extract primary key
            pk_match = re.search(r"PRIMARY KEY: (\w+)", chunk)
            if pk_match:
                schema_info['primary_keys'][table_name] = pk_match.group(1)

            # Extract columns
            column_matches = re.findall(r"(\w+)\s+([A-Z]+(?:\([^)]+\))?)", chunk)

            for col_name, col_type in column_matches:
                # Skip common id columns for keywords
                if col_name != 'id' and not col_name.endswith('_id'):
                    schema_info['keywords'].add(col_name.lower())

                # Extract base data type
                base_type = col_type.split('(')[0].strip()
                schema_info['data_types'].add(base_type.lower())

                schema_info['tables'][table_name]['columns'].append({
                    'name': col_name,
                    'type': col_type
                })

                # Track common columns across tables for join detection
                schema_info['common_columns'].add(col_name)

            # Extract foreign key relationships
            fk_matches = re.findall(r"FOREIGN KEY: (\w+) REFERENCES (\w+)\((\w+)\)", chunk)
            for fk_col, ref_table, ref_col in fk_matches:
                schema_info['relationships'].append({
                    'table': table_name,
                    'fk_column': fk_col,
                    'ref_table': ref_table,
                    'ref_column': ref_col
                })
                if table_name not in schema_info['foreign_keys']:
                    schema_info['foreign_keys'][table_name] = []
                schema_info['foreign_keys'][table_name].append(fk_col)

    # Extract enum values as keywords
    enum_matches = re.findall(r"ENUM\('([^']+)'(,\s*'[^']+')*\)", ' '.join(schema_chunks))
    for enum_match in enum_matches:
        full_match = enum_match[0]
        for value in re.findall(r"'([^']+)'", full_match):
            schema_info['keywords'].add(value.lower())

    # Filter out SQL-specific stopwords and very short keywords
    schema_info['keywords'] = {k for k in schema_info['keywords']
                                if k not in sql_stopwords
                                and k not in english_stopwords
                                and len(k) > 2}

    return schema_info

# ----------------------------
# Enhanced Keyword Extraction
# ----------------------------
def extract_keywords_from_question(question, schema_info):
    """Extract keywords from the question based on schema information"""
    question_lower = question.lower()
    words = set(re.findall(r'\b\w+\b', question_lower))

    # Filter out stopwords and short words
    english_stopwords = set(stopwords.words('english'))
    sql_keywords = {'select', 'from', 'where', 'group', 'by', 'having', 'order',
                   'join', 'left', 'right', 'inner', 'outer', 'on', 'and', 'or',
                   'not', 'in', 'between', 'like', 'is', 'null', 'as', 'limit'}

    filterable_words = words - english_stopwords - sql_keywords

    # Extract keywords matching schema elements
    table_keywords = {word for word in filterable_words if word in {t.lower() for t in schema_info['tables']}}
    column_keywords = {word for word in filterable_words if word in {col['name'].lower() for table in schema_info['tables'].values() for col in table['columns']}}
    value_keywords = {word for word in filterable_words if word in schema_info['keywords']}

    # Look for potential aggregations or operations
    operation_keywords = {'count', 'sum', 'average', 'avg', 'max', 'min', 'total',
                         'number', 'list', 'find', 'rank', 'top', 'bottom', 'most',
                         'least', 'highest', 'lowest', 'calculate', 'show'}
    op_keywords = {word for word in filterable_words if word in operation_keywords}

    # Time-related keywords
    time_keywords = {'date', 'time', 'year', 'month', 'day', 'week', 'hour',
                    'recent', 'latest', 'oldest', 'current', 'previous', 'next',
                    'last', 'first', 'quarter', 'annual'}
    temporal_keywords = {word for word in filterable_words if word in time_keywords}

    # Comparison keywords
    comparison_keywords = {'greater', 'less', 'equal', 'before', 'after', 'between'}
    comp_keywords = {word for word in filterable_words if word in comparison_keywords}

    # Combine all relevant keywords
    all_keywords = (table_keywords | column_keywords | value_keywords |
                   op_keywords | temporal_keywords | comp_keywords)

    # If we didn't find many keywords, include all non-stopwords as fallback
    if len(all_keywords) < 3:
        all_keywords = {word for word in filterable_words if len(word) > 3}

    return list(all_keywords)

def extract_table_column_references(question, schema_info):
    """Extract specific table and column references from the question"""
    question_lower = question.lower()
    table_references = []
    column_references = []

    # Check for direct table references
    for table_name in schema_info['tables']:
        table_lower = table_name.lower()
        # Exact match
        if re.search(rf'\b{table_lower}\b', question_lower):
            table_references.append(table_name)
        # Check singular form too
        elif table_lower.endswith('s') and re.search(rf'\b{table_lower[:-1]}\b', question_lower):
            table_references.append(table_name)

    # Check for column references
    for table_name, table_info in schema_info['tables'].items():
        for column in table_info['columns']:
            col_name = column['name']
            col_lower = col_name.lower()
            if re.search(rf'\b{col_lower}\b', question_lower) and col_lower not in ['id', 'created_at', 'updated_at']:
                column_references.append((table_name, col_name))

    return table_references, column_references

def get_related_tables(table_references, schema_info):
    """Get tables related to the referenced tables through foreign keys"""
    all_tables = set(table_references)

    # Add related tables through foreign keys (recursively)
    changed = True
    while changed:
        changed = False
        for rel in schema_info['relationships']:
            if rel['table'] in all_tables and rel['ref_table'] not in all_tables:
                all_tables.add(rel['ref_table'])
                changed = True
            elif rel['ref_table'] in all_tables and rel['table'] not in all_tables:
                all_tables.add(rel['table'])
                changed = True

    return list(all_tables)

# ----------------------------
# Enhanced Query Expansion
# ----------------------------
def expand_query(question, schema_info):
    """Expand the query with schema-related terms"""
    # Extract keywords from question
    keywords = extract_keywords_from_question(question, schema_info)

    # Find WordNet synonyms for important keywords
    expanded_terms = set()
    for word in keywords:
        # Skip very short words and numbers
        if len(word) <= 3 or word.isdigit():
            continue

        expanded_terms.add(word)
        synsets = wordnet.synsets(word)
        for synset in synsets[:2]:  # Take only first 2 synsets to avoid too much noise
            for lemma in synset.lemmas()[:2]:  # Take only first 2 lemmas
                expanded_term = lemma.name().replace('_', ' ')
                # Only add synonyms that aren't already in our keywords
                if expanded_term not in keywords:
                    expanded_terms.add(expanded_term)

    # Add common join conditions based on foreign keys
    table_refs, _ = extract_table_column_references(question, schema_info)
    related_tables = get_related_tables(table_refs, schema_info)

    for table in related_tables:
        if table in schema_info['foreign_keys']:
            for fk in schema_info['foreign_keys'][table]:
                expanded_terms.add(f"{table}.{fk}")

    # Combine with original question
    expanded_query = question + " " + " ".join(list(expanded_terms - set(keywords)))
    return expanded_query

# ----------------------------
# Enhanced Context Retrieval
# ----------------------------
def get_context(question, schema_chunks, schema_info, top_k=3, use_query_expansion=True):
    """Enhanced RAG retrieval function with dynamic keyword extraction"""
    # TECHNIQUE 1: Query expansion
    if use_query_expansion:
        expanded_question = expand_query(question, schema_info)
    else:
        expanded_question = question

    # TECHNIQUE 2: Dynamic keyword extraction
    keywords = extract_keywords_from_question(expanded_question, schema_info)
    table_refs, column_refs = extract_table_column_references(expanded_question, schema_info)

    # Get related tables through foreign keys
    all_table_refs = get_related_tables(table_refs, schema_info)

    # First-pass filtering: keyword-based with dynamic keywords
    all_keywords = set(keywords)
    for table in all_table_refs:
        all_keywords.add(table.lower())
    for table, column in column_refs:
        all_keywords.add(column.lower())

    # Filter schema chunks based on relevance to keywords
    filtered_chunks = []
    chunk_scores = []

    for chunk in schema_chunks:
        chunk_lower = chunk.lower()

        # Count keyword matches
        keyword_matches = sum(1 for keyword in all_keywords if re.search(rf'\b{keyword}\b', chunk_lower))

        # Check for table references - give higher weight
        table_matches = sum(5 for table in all_table_refs if f"table: {table.lower()}" in chunk_lower)

        # Check for column references - give medium weight
        column_matches = sum(3 for _, column in column_refs if re.search(rf'\b{column.lower()}\b', chunk_lower))

        # Calculate total relevance score
        total_score = keyword_matches + table_matches + column_matches

        if total_score > 0:
            filtered_chunks.append(chunk)
            chunk_scores.append(total_score)

    # If keyword filtering is too restrictive, fallback to all schemas
    if len(filtered_chunks) < 2:
        filtered_chunks = schema_chunks
        chunk_scores = [1] * len(schema_chunks)

    # Second-pass: semantic search
    chunk_embeddings = embedding_model.encode(filtered_chunks)
    question_embedding = embedding_model.encode([expanded_question])
    semantic_similarities = cosine_similarity(question_embedding, chunk_embeddings)[0]

    # TECHNIQUE 3: Re-ranking with combined score
    max_keyword_score = max(chunk_scores) if chunk_scores else 1
    normalized_keyword_scores = [score / max_keyword_score for score in chunk_scores]

    # Combine semantic similarity with keyword matching (70% semantic, 30% keyword)
    combined_scores = [0.7 * sem_score + 0.3 * key_score
                      for sem_score, key_score in zip(semantic_similarities, normalized_keyword_scores)]

    # Get top-k chunks based on combined score
    ranked_indices = np.argsort(combined_scores)[::-1]
    top_chunks = [filtered_chunks[i] for i in ranked_indices[:top_k]]

    # Add schemas for any missing but referenced tables
    referenced_tables = set(all_table_refs)
    included_tables = set()

    for chunk in top_chunks:
        for table in referenced_tables:
            if f"Table: {table}" in chunk:
                included_tables.add(table)

    missing_tables = referenced_tables - included_tables
    if missing_tables:
        for table in missing_tables:
            for chunk in schema_chunks:
                if f"Table: {table}" in chunk and chunk not in top_chunks:
                    top_chunks.append(chunk)
                    break

    return "\n".join(top_chunks)
```

```python
# ----------------------------
# Enhanced Few-shot Examples
# ----------------------------
few_shot_examples = [
    {
        "question": "List all customers who made a purchase last month",
        "sql": "SELECT DISTINCT c.* FROM customers c JOIN sales s ON c.id = s.customer_id WHERE s.sale_date >= DATE_SUB(CURRENT_DATE(), INTERVAL 1 MONTH);"
    },
    {
        "question": "What's the total revenue by product?",
        "sql": "SELECT p.name, SUM(s.total_amount) as total_revenue FROM products p JOIN sales s ON p.id = s.product_id GROUP BY p.id, p.name ORDER BY total_revenue DESC;"
    },
    {
        "question": "Find customers with more than 3 interactions but no sales",
        "sql": """SELECT c.*
FROM customers c
JOIN (
    SELECT customer_id, COUNT(*) as interaction_count
    FROM interactions
    GROUP BY customer_id
    HAVING COUNT(*) > 3
) i ON c.id = i.customer_id
LEFT JOIN sales s ON c.id = s.customer_id
WHERE s.id IS NULL;"""
    },
    {
        "question": "Show the average opportunity value by stage",
        "sql": "SELECT opportunity_stage, AVG(value) as avg_value FROM sales_opportunities GROUP BY opportunity_stage ORDER BY avg_value DESC;"
    }
]

# ----------------------------
# Enhanced SQL Validation
# ----------------------------
def validate_and_fix_sql(sql, schema_info):
    """Validate SQL syntax and attempt to fix common errors"""
    try:
        parsed = sqlparse.parse(sql)
        if not parsed:
            raise ValueError("SQL parsing failed.")

        # Additional validation checks
        sql_lower = sql.lower()

        # Check for proper table references
        for table in schema_info['tables']:
            table_lower = table.lower()
            if (re.search(rf'\b{table_lower}\b', sql_lower) and
                not re.search(rf'(from|join)\s+{table_lower}\b', sql_lower)):
                sql += f" /* Warning: Table '{table}' is referenced but not properly used with FROM or JOIN */"

        # Check for basic SQL structure
        if not re.search(r'^\s*select\b', sql_lower, re.IGNORECASE):
            sql = "SELECT " + sql

        # Check for missing join conditions
        if len(re.findall(r'\bjoin\b', sql_lower, re.IGNORECASE)) > 0:
            if not re.search(r'\bon\b', sql_lower, re.IGNORECASE):
                sql += " /* Warning: JOIN without ON condition */"

        return True, sql

    except Exception as e:
        error_msg = str(e)
        print(f"SQL syntax error: {error_msg}")

        # Try to fix common errors
        fixed_sql = sql

        # Missing semicolon
        if not fixed_sql.strip().endswith(';'):
            fixed_sql = fixed_sql + ';'

        # Unclosed parentheses
        open_count = fixed_sql.count('(')
        close_count = fixed_sql.count(')')
        if open_count > close_count:
            fixed_sql = fixed_sql + ')' * (open_count - close_count)

        # Try parsing again with fixes
        try:
            parsed = sqlparse.parse(fixed_sql)
            if parsed:
                return True, fixed_sql
        except:
            pass

        return False, sql

# ----------------------------
# Enhanced SQL Generation
# ----------------------------
def generate_sql(question, max_attempts=2):
    """Generate SQL with enhanced RAG and validation"""
    # Analyze schema once
    schema_info = analyze_schema(schema_chunks)

    # Get enhanced context with dynamic keyword extraction
    context = get_context(question, schema_chunks, schema_info, top_k=4)

    # Log the extracted context and keywords for debugging
    print(f"\nExtracted keywords: {extract_keywords_from_question(question, schema_info)}")
    table_refs, column_refs = extract_table_column_references(question, schema_info)
    print(f"Table references: {table_refs}")
    print(f"Column references: {column_refs}")
    print(f"Related tables: {get_related_tables(table_refs, schema_info)}")

    # Construct few-shot examples
    few_shot_text = ""
    for example in few_shot_examples:
        few_shot_text += f"Q: {example['question']}\nSQL: {example['sql']}\n\n"

    # First attempt with detailed prompt
    prompt = f"""You are an expert SQL query constructor. Generate a valid SQL query for the given question based on the database schema.

Database Schema:
{context}

Important Notes:
1. Always include proper JOIN conditions when using multiple tables
2. Use table aliases for clarity (e.g., customers c)
3. Include all required columns in SELECT
4. Use appropriate WHERE conditions based on the question
5. End with a semicolon

Examples:
{few_shot_text}
Question: {question}

SQL:"""

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    output = model.generate(**inputs, max_new_tokens=256, do_sample=False)
    sql = tokenizer.decode(output[0], skip_special_tokens=True)

    # Clean up the generated SQL
    sql = sql.split("SQL:")[-1].strip()
    sql = re.sub(r'\n+', ' ', sql)  # Replace newlines with spaces
    sql = re.sub(r'\s+', ' ', sql)  # Collapse multiple spaces
    sql = sql.split(';')[0] + ';'  # Ensure exactly one semicolon at end

    # Validate and potentially fix the SQL
    valid, sql = validate_and_fix_sql(sql, schema_info)

    # If not valid and we have attempts left, try again with a more structured prompt
    attempt = 1
    while not valid and attempt < max_attempts:
        attempt += 1
        print(f"Retrying SQL generation (attempt {attempt})...")

        # More structured prompt for the retry
        prompt = f"""You are an expert SQL query constructor. Generate a valid SQL query following these rules:
1. Start with SELECT [columns] FROM [table]
2. Use proper JOIN syntax with ON conditions when needed
3. Include WHERE clauses for filtering as needed
4. Use GROUP BY for aggregations
5. End with a semicolon

Database Schema:
{context}

Question: {question}

SQL:"""

        inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
        output = model.generate(**inputs, max_new_tokens=256, do_sample=True,
                               temperature=0.7)  # Slightly randomize to get different output
        sql = tokenizer.decode(output[0], skip_special_tokens=True).split("SQL:")[-1].strip()

        # Clean up and validate again
        sql = re.sub(r'\n+', ' ', sql)
        sql = re.sub(r'\s+', ' ', sql)
        sql = sql.split(';')[0] + ';'
        valid, sql = validate_and_fix_sql(sql, schema_info)

    return sql
```
